# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import math
```

# Set A [7A1] – Exam Scores Dataset

Create Dataset

```python
scores = np.array([93, 83, 69, 62, 75, 93, 73, 77, 65, 65, 78, 90, 94, 78, 57, 76, 56, 78, 84, 92])

print(scores)
```

#1. Basic Statistics

```python
mean_score = np.mean(scores)

# Population Variance & Std
pop_variance = np.var(scores)
pop_std = np.std(scores)

# Sample Variance
sample_variance = np.var(scores, ddof=1)

print("Mean:", mean_score)
print("Population Variance:", pop_variance)
print("Sample Variance:", sample_variance)
print("Standard Deviation:", pop_std)
```

Explanation of Population vs Sample Variance

```python
print("Population variance divides by N.")
print("Sample variance divides by N-1.")
print("Sample variance is slightly larger because it corrects estimation bias.")
```

#2. Random Number Generation

```python
np.random.seed(42)

generated_scores = np.random.normal(
    loc=mean_score,
    scale=pop_std,
    size=20
)

print(generated_scores)
```

Histogram Comparison

```python
plt.figure(figsize=(12,5))

# Real Data Histogram
plt.subplot(1,2,1)
plt.hist(scores, bins=6)
plt.title("Real Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

# Generated Data Histogram
plt.subplot(1,2,2)
plt.hist(generated_scores, bins=6)
plt.title("Generated Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#3. Monte Carlo Simulation

```python
simulation = np.random.normal(
    loc=mean_score,
    scale=pop_std,
    size=100000
)

probability = np.mean(simulation >= 75)

print("Probability of scoring 75 or above:", probability)
```

#4. Hypothesis Testing

```python
hypothesized_mean = (96%50)+50 # Last 3 Digit of my ID is 096.

# One Sample t-test

t_stat, p_value = stats.ttest_1samp(scores, hypothesized_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

Visualization and Conclusion

```python
plt.figure(figsize=(8,5))

plt.hist(scores, bins=6)
plt.axvline(np.mean(scores), linestyle='--', label='Sample Mean')
plt.axvline(hypothesized_mean, linestyle='--', label='Hypothesized Mean')

plt.title("Hypothesis Testing Visualization")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()

plt.show()

print()
print()
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
    print("The average exam score is significantly different from 96.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("The average exam score is not significantly different from 96.")
```

# Set A [7A2] – Customer Arrival Dataset

Create Dataset

```python
arrivals = np.array([8, 12, 10, 15, 9, 13, 11, 14, 7, 16, 10, 12, 9, 15, 11, 13, 14, 8, 12, 10])

print(arrivals)
```

#1. Basic Statistics

```python
mean_arrival = np.mean(arrivals)

# Population Variance & Std
pop_variance = np.var(arrivals)
pop_std = np.std(arrivals)

# Sample Variance
sample_variance = np.var(arrivals, ddof=1)

print("Mean:", mean_arrival)
print("Population Variance:", pop_variance)
print("Sample Variance:", sample_variance)
print("Standard Deviation:", pop_std)
```

Explain the Difference

```python
print("Population variance uses N.")
print("Sample variance uses N-1.")
print("Sample variance is slightly larger because it estimates from a sample.")
```

#2. Random Number Generation

```python
np.random.seed(42)

simulated_arrivals = np.random.uniform(
    low=min(arrivals),
    high=max(arrivals),
    size=30
)

print(simulated_arrivals)
```

Histogram Comparison

```python
plt.figure(figsize=(12,5))

# Real Data
plt.subplot(1,2,1)
plt.hist(arrivals, bins=6)
plt.title("Real Customer Arrivals")
plt.xlabel("Arrivals")
plt.ylabel("Frequency")

# Simulated Data
plt.subplot(1,2,2)
plt.hist(simulated_arrivals, bins=6)
plt.title("Simulated Customer Arrivals")
plt.xlabel("Arrivals")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#3. Monte Carlo Simulation

```python
mc_simulation = np.random.uniform(
    low=min(arrivals),
    high=max(arrivals),
    size=100000
)

probability = np.mean(mc_simulation > 15)

print("Probability of more than 15 customers:", probability)
```

#4. Hypothesis Testing

```python
hypothesized_mean = (96%10)+10 # Last 3 Digit of my ID is 096.

# One Sample t-test

t_stat, p_value = stats.ttest_1samp(arrivals, hypothesized_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

Visualization and Conclusion

```python
plt.figure(figsize=(8,5))

plt.hist(arrivals, bins=6)
plt.axvline(np.mean(arrivals), linestyle='--', label='Sample Mean')
plt.axvline(hypothesized_mean, linestyle='--', label='Hypothesized Mean')

plt.title("Customer Arrival Hypothesis Test")
plt.xlabel("Daily Arrivals")
plt.ylabel("Frequency")
plt.legend()

plt.show()

print()
print()

alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
    print("The mean customer arrival is significantly different from 16.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("The mean customer arrival is not significantly different from 16.")
```
