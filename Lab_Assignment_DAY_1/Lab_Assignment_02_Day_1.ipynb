```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# a. Create two vectors, each containing 15 random floats of your choice.

vec_a = np.random.rand(15)
vec_b = np.random.rand(15)

print("Vector A:", vec_a)
print("Vector B:", vec_b)


indices = np.arange(vec_a.size)
print("Indices:", indices)

# b. Plot these vectors on the same graph using matplotlib to compare how they look.

plt.plot(indices, vec_a, marker='^', linestyle='-', color='blue', label="Vector A")
plt.plot(indices, vec_b, marker='d', linestyle=':', color='orange', label="Vector B")

# c. Label the axes appropriately and give the plot a title.
plt.xlabel("Index Position")
plt.ylabel("Random Value")
plt.title("Comparison of Two Random Vectors")
plt.legend()
plt.grid(alpha=0.6)

plt.show()
```
