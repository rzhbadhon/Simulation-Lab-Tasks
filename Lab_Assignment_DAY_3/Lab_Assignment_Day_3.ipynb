```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# Part 1 —  Data Generation

```python

np.random.seed(42)
regular_wait = np.random.uniform(2, 12, 3000)

busy_wait = np.random.normal(8, 2, 3000)

busy_wait = np.where(busy_wait < 0, 0, busy_wait)
```

#  Part 2 — Visualization

```python
plt.figure(figsize=(12,5))

# Regular Day
plt.subplot(1,2,1)
sns.histplot(regular_wait, bins=30)
plt.title("Regular Day - Uniform Distribution")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")

# Busy Day
plt.subplot(1,2,2)
sns.histplot(busy_wait, bins=30)
plt.title("Busy Day - Normal Distribution")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

# Part 3 — Statistical Analysis

```python
# Regular Day
print("Regular Day Statistics:")
print("Mean:", np.mean(regular_wait))
print("Median:", np.median(regular_wait))
print("Standard Deviation:", np.std(regular_wait))
print("Maximum:", np.max(regular_wait))

# Busy Day
print("\nBusy Day Statistics:")
print("Mean:", np.mean(busy_wait))
print("Median:", np.median(busy_wait))
print("Standard Deviation:", np.std(busy_wait))
print("Maximum:", np.max(busy_wait))
```
