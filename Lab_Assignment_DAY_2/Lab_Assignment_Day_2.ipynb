```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A = np.array([[3+2, 8+1], [2*3, 8+2]])
print(A)
```

```python
# 1. shape
dims = np.shape(A)
print("Dimensions: ", dims)

# 2. Determinant
det_val = np.linalg.det(A)
print("Determinant:", det_val)

# 3. Inverse matrix check
if det_val != 0:
    inv_mat = np.linalg.inv(A)
else:
    inv_mat = "Not possible"

print("Inverse:", inv_mat)

# Rank of matrix
mat_rank = np.linalg.matrix_rank(A)
print("Rank:", mat_rank)

# Eigenvalues and eigenvectors
eig_vals, eig_vecs = np.linalg.eig(A)
print("Eigenvalues:", eig_vals)
print("Eigenvectors:\n", eig_vecs)
```

```python
B = np.array([[3+2+1, 8+1],
              [2*3,   8+2]])
print(B)
```

```python
# 1. shape
dims = np.shape(B)
print("Dimensions: ", dims)

# 2. Determinant
det_val = np.linalg.det(B)
print("Determinant:", det_val)

# 3. Inverse matrix check
if det_val != 0:
    inv_mat = np.linalg.inv(B)
else:
    inv_mat = "Not possible"

print("Inverse:", inv_mat)

# Rank of matrix
mat_rank = np.linalg.matrix_rank(B)
print("Rank:", mat_rank)

# Eigenvalues and eigenvectors
eig_vals, eig_vecs = np.linalg.eig(B)
print("Eigenvalues:", eig_vals)
print("Eigenvectors:\n", eig_vecs)
```

```python
# Determinant change: Compare det_val vs det_B. The difference comes from altering one entry, which shifts the balance of multiplication terms in the determinant formula.

# Rank: If the rows/columns remain linearly independent, rank stays 2. If the change makes them dependent, rank drops.

# Eigenvalues: They shift because the characteristic polynomial changes when any entry is modified.

# Invertibility: If determinant moves closer to 0, matrix B is harder to invert (numerically unstable). If determinant increases in magnitude, inversion is easier.
```
