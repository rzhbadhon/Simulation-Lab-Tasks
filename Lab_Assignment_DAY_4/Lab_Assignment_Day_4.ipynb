```python
import random
import numpy as np
import math
```

```python
# last two digits of my ID
R = 83

range_max = R / 10

radius = R / 10

full_name = "Razibul Hasan Badhon"

letters_count = len(full_name.replace(" ", ""))

min_points = 200 * letters_count

print("Range Used: [0,", range_max, "]")
print("Minimum Points:", min_points)
```

```python
def is_inside_circle(x, y, radius):
    if x**2 + y**2 <= radius**2:
        return True
    else:
        return False
```

```python
def run_simulation(total_points):

    inside = 0
    outside = 0

    half = total_points // 2

    # First half using random.uniform()
    for i in range(half):
        x = random.uniform(0, range_max)
        y = random.uniform(0, range_max)

        if is_inside_circle(x, y, radius):
            inside += 1
        else:
            outside += 1

    # Second half using numpy.random.rand()
    for i in range(total_points - half):
        x = np.random.rand() * range_max
        y = np.random.rand() * range_max

        if is_inside_circle(x, y, radius):
            inside += 1
        else:
            outside += 1

    # π estimation
    pi_est = 4 * inside / total_points

    error = abs(math.pi - pi_est)

    return inside, outside, pi_est, error
```

```python
print("Range Used | Total Points | Points Inside | Estimated π | Error from math.pi")
print("--------------------------------------------------------------------------")

for i in range(1, 6):
    total = min_points * i

    inside, outside, pi_est, error = run_simulation(total)

    print(f"[0,{range_max}] | {total} | {inside} | {round(pi_est,6)} | {round(error,6)}")
```
